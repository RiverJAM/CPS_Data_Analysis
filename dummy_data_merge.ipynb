```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import the csv files for school progress reports and the teacher salaries.  merge into a single data frame

salaries = pd.read_csv('Resources/salaries_06302020_high_school_teachers.csv')
schools = pd.read_csv('Resources/Chicago_Public_Schools_-_School_Progress_Reports_SY1819_cleaned.csv', encoding='iso-8859-1')
combined_df = pd.merge(salaries, schools, on = "School_ID", how = "left")
combined_df.columns.tolist()
```

```python
combined_df
```

```python
x_values = combined_df.groupby(combined_df['Department']).mean()['FTE Annual Salary']
y_values = combined_df.groupby(combined_df['Department']).mean()['Attainment_SAT_Grade_11_School_Pct']
type(y_values)
```
